```python
from trie import Node, Trie
from graphviz import Digraph
```

```python
trie = Trie()
trie.build("aabbabd")
```

```python
def draw_tree(root):
    tree = Digraph()
    def build(node):
        tree.node(label=str(node.get_val()), name=str(id(node)))
        if node.parent != None:
            tree.edge(str(id(node.parent)), str(id(node)))
        for char in node.children:
            build(node.get_child(char))

    build(root)
    return tree
```

```python
draw_tree(trie.root)
```

```python
trie2 = Trie()
trie2.build("abcbccd")
```

```python
draw_tree(trie2.root)
```
